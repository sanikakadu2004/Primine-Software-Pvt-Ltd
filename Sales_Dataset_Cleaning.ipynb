```python
# Data Preprocessing
# Read CSV file
df = pd.read_csv("Sales_Dataset.csv")

# Check basic info
print(df.info())
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Date columns
date_cols = ["OrderDate", "ShipmentDate", "DeliveryDate"]
# Convert & format dates
for col in date_cols:
    # First attempt (general parsing)
    parsed = pd.to_datetime(df[col], errors="coerce", dayfirst=True)
    # Second attempt for still-missing values
    parsed2 = pd.to_datetime(
        df[col],
        errors="coerce",
        format="mixed"
    )
    # Combine both
    df[col] = parsed.fillna(parsed2)
    # Final format
    df[col] = df[col].dt.strftime("%d-%m-%Y")

# Fill missing values with "NA"
df = df.fillna("NA")
df
```

```python
# Product Cleaning
df["Product"] = (
    df["Product"]
    .str.lower()
    .str.replace("-", " ", regex=False)
    .str.replace("_", " ", regex=False)
    # add space between letters and numbers
    .str.replace(r"([a-zA-Z])(\d)", r"\1 \2", regex=True)
    # add space between words like mousewireless
    .str.replace(r"(mouse)(wireless)", r"\1 \2", regex=True)
    .str.strip()
    .str.title()
)

df["CustomerName"] = (
    df["CustomerName"]
    .astype(str)
    .str.replace("_", " ", regex=False) # underscores → space
    .str.replace("-", " ", regex=False) # hyphens → space
    # add space between letters & numbers
    .str.replace(r"([a-zA-Z])(\d)", r"\1 \2", regex=True)
    # split CamelCase words
    .str.replace(r"([a-z])([A-Z])", r"\1 \2", regex=True)
    # normalize multiple spaces
    .str.replace(r"\s+", " ", regex=True)
    .str.strip()
    .str.title()
)

# Use a generic suffix-based standardization
import re
def clean_customer_name(name):
    name = str(name).lower()
    # replace separators
    name = name.replace("_", " ").replace("-", " ")
    # split camel case
    name = re.sub(r"([a-z])([A-Z])", r"\1 \2", name)
    # normalize multiple spaces
    name = re.sub(r"\s+", " ", name).strip()
    # merge common business compounds
    name = re.sub(r"\binfo\s+tech\b", "infotech", name)
    name = re.sub(r"\btech\s+solutions\b", "techsolutions", name)
    return name.title()
df["CustomerName"] = df["CustomerName"].apply(clean_customer_name)
df

import re
def clean_product(product):
    p = str(product).strip()
    # Replace separators
    p = re.sub(r"[_\-]", " ", p)
    # Split CamelCase
    p = re.sub(r"([a-z])([A-Z])", r"\1 \2", p)
    # Split letters & numbers
    p = re.sub(r"([A-Za-z])(\d)", r"\1 \2", p)
    p = re.sub(r"(\d)([A-Za-z])", r"\1 \2", p)
    # Normalize
    p = p.lower()
    # Common tech suffix separation (GENERAL, extensible)
    p = re.sub(r"(usb|ssd|hdd|ram|gb|tb|inch|in)", r" \1", p)
    # Clean spaces
    p = re.sub(r"\s+", " ", p).strip()
    return p.title()
df["Product"] = df["Product"].apply(clean_product)
df
```

```python
# Calculate mean excluding zeros and round to 1 decimal
qty_mean = round(df.loc[df["Quantity"] != 0, "Quantity"].mean(), 1)
# # Fill quantity=0 with mean
df["Quantity"] = df["Quantity"].replace(0, qty_mean)
df
```

```python
# Convert 'UnitPrice' to numeric, coercing 'NA' to NaN
df["UnitPrice"] = pd.to_numeric(df["UnitPrice"], errors='coerce')
# Fill missing UnitPrice with median
df["UnitPrice"].fillna(df["UnitPrice"].median(), inplace=True)
df
```
