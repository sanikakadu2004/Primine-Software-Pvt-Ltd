```python
# Data Preprocessing
# Read CSV file
df = pd.read_csv("Clean_Crm_Dataset.csv")

# Convert column names to CAPITAL
df.columns = [col.upper() for col in df.columns]
print(df.columns)
```

```python
# Remove Duplicate Records
df = df.drop_duplicates()

# String Cleaning
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Standardize Text Data
df["STATE"] = df["STATE"].str.upper()
df["COUNTRY"] = df["COUNTRY"].str.upper()

# CREATED_DATE → YYYY/MM/DD
df["CREATED_DATE"] = pd.to_datetime(
    df["CREATED_DATE"], errors="coerce"
).dt.strftime("%d-%m-%Y")

# LAST_CONTACTED → DD-MM-YYYY
df["LAST_CONTACTED"] = pd.to_datetime(
    df["LAST_CONTACTED"], dayfirst=True, errors="coerce"
).dt.strftime("%d-%m-%Y")

# Fill missing values with "NA"
df = df.fillna("NA")
df
```

```python
# Save preprocessed file
df.to_csv("Preprocessed_CRM_Dataset.csv", index=False)
# Download preprocessed file
files.download("Preprocessed_CRM_Dataset.csv")
```
